```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../dataset/sample_students.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
numeric_cols = ['attendance_percent', 'study_hours', 'internal_marks', 'assignments_submitted']

plt.figure(figsize=(12,8))
for i, col in enumerate(numeric_cols):
    plt.subplot(2,2,i+1)
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.countplot(x=df['target'])
plt.title("Target Distribution (Pass/Fail)")
plt.show()
```

```python
sns.pairplot(df, hue="target")
plt.show()
```

```python
print("INSIGHTS:")

print("\n1. Class Balance:")
print(df['target'].value_counts())

print("\n2. Correlation Insights:")
cor = df.corr()['target'].sort_values(ascending=False)
print(cor)

print("\n3. Feature Importance Suggestions:")
print("""
- Internal marks show strongest correlation with passing outcome.
- Attendance also contributes strongly.
- Assignments submitted and study hours show moderate effect.
- Activities participation has small but positive effect.
""")
```
